```python
import numpy as np
import matplotlib 
import matplotlib.pyplot as pylab
pi = np.pi
import csv
import pandas as pd
from pandas import HDFStore, DataFrame
# hdf =HDFStore('storage.h5')
# from pandas import read_hdf
import math
%matplotlib inline
```

```python
def cylinder(radius, height):
    face = pi*radius**2
    sides = 2*pi*height
    surface = sides+face*2
    volume = face * height
    return volume
```

```python
def float_ur_boat(input_number):
    output = float("{0:.3f}".format(input_number))
    return output
```

```python
def square_pack_rad(active_diam, fcage_thickn, ics_thickn, DB_length):
    r         = active_diam + fcage_thickn + ics_thickn
    DB_area   = DB_length**2
    pack_num  = math.floor( (pi*(r**2)) / DB_area) 
    DB_Bi     = 0.0700 * pack_num
    DB_Co     = 0.0060 * pack_num
    DB_K      = 14.500 * pack_num
    DB_Tl     = 0.0104 * pack_num
    
    return DB_Bi, DB_Co, DB_K, DB_Tl
```

```python
def volume_suite(active_diam ,  active_length, 
                 fcage_thickn,  ics_thickn,
                 Out1_length ,  Out2_length,
                 ROP1_length ,  ROP2_length,
                 vessel_thickn):
    
    
    active_diam   = float_ur_boat(active_diam)
    active_length = float_ur_boat(active_length)
    active_rad    = float_ur_boat(active_diam/2)


    fcage_thickn  = float_ur_boat(fcage_thickn)
    fcage_rad     = float_ur_boat(active_rad + fcage_thickn)
    fcage_length  = float_ur_boat(active_length)


    ics_thickn    = float_ur_boat(ics_thickn)
    ics_rad       = float_ur_boat(fcage_rad + ics_thickn)
    ics_length    = float_ur_boat(active_length)
    
    Out1_length   = float_ur_boat(Out1_length)
    Out1_rad      = float_ur_boat(ics_rad)
    Out2_length   = float_ur_boat(Out2_length)
    Out2_rad      = float_ur_boat(ics_rad)
    
    ROP1_length   = float_ur_boat(ROP1_length)
    ROP1_rad      = float_ur_boat(ics_rad)
    ROP2_length   = float_ur_boat(ROP2_length)
    ROP2_rad      = float_ur_boat(ics_rad)
    
    det_length    = .01
    det_rad       = active_rad


    vessel_thickn = float_ur_boat(vessel_thickn)
    vessel_rad    = float_ur_boat(ics_rad + vessel_thickn)
    vessel_length = float_ur_boat(active_length + Out1_length + Out2_length + 2*vessel_thickn)


    active_vol    = float_ur_boat(cylinder(active_rad, active_length))
    fcage_vol     = float_ur_boat(cylinder(fcage_rad , fcage_length)) - active_vol
    ics_vol       = float_ur_boat(cylinder(ics_rad   , ics_length)) - fcage_vol - active_vol
    Out1_vol      = float_ur_boat(cylinder(Out1_rad   , Out1_length))
    Out2_vol      = float_ur_boat(cylinder(Out2_rad   , Out2_length))
    ROP1_vol      = float_ur_boat(cylinder(ROP1_rad   , ROP1_length))
    ROP2_vol      = float_ur_boat(cylinder(ROP2_rad   , ROP2_length))
    vessel_vol    = float_ur_boat(cylinder(vessel_rad, vessel_length)) - ics_vol - fcage_vol - active_vol - Out1_vol - Out2_vol
    det_vol       = float_ur_boat(cylinder(det_rad, det_length))
    
    return (active_vol, fcage_vol, 
            ics_vol   , Out1_vol,
            Out2_vol  , ROP1_vol,
            ROP2_vol  , vessel_vol, 
            det_vol)
```

```python
def radioactivity(active_diam ,  active_length, 
                  fcage_thickn,  ics_thickn,
                  Out1_length ,  Out2_length, 
                  ROP1_length , ROP2_length,
                  vessel_thickn):

    (Act , FC, 
     ICS , Out1, 
     Out2, ROP1,
     ROP2, Ves, 
     Det) = volume_suite(active_diam ,  active_length, 
                         fcage_thickn,  ics_thickn,
                         Out1_length ,  Out2_length,
                         ROP1_length , ROP2_length,
                         vessel_thickn)
    
    
    FC_HDPE         = 1 * FC
    FC_Tef          = 0.2543 * FC
    FC_Cu           = 0.0615 * FC
    
    FC_HDPE_Mass    = 0.00095 * FC_HDPE
    FC_Tef_Mass     = 0.00220 * FC_Tef
    FC_Cu_Mass      = 0.00894 * FC_Cu
    
    ICS_Mass        = 0.00894 * ICS
    
    Out1_Mass       = 0.00894 * Out1
    Out2_Mass       = 0.00894 * Out2
    
    ROP1_Mass       = 0.00095 * ROP1
    ROP2_Mass       = 0.00095 * ROP2
    
    Ves_Mass        = 0.00790 * Ves
    Det_Mass        = .001420 * Det
      
    FC_HDPE_Bi_Rad  = 0.06200 * FC_HDPE_Mass
    FC_Tef_Bi_Rad   = 1.80000 * FC_Tef_Mass
    FC_Cu_Bi_Rad    = 0.01200 * FC_Cu_Mass
#     FC_Bi_Rad       = float_ur_boat(FC_HDPE_Bi_Rad + FC_Tef_Bi_Rad + FC_Cu_Bi_Rad )
    FC_Bi_Rad       = float_ur_boat(FC_HDPE_Bi_Rad)
    
    ICS_Bi_Rad      = float_ur_boat(0.01200 * ICS_Mass)
    
    Out1_Bi_Rad     = float_ur_boat(0.01200 * Out1_Mass)
    Out2_Bi_Rad     = float_ur_boat(0.01200 * Out2_Mass)
    
    ROP1_Bi_Rad     = float_ur_boat(0.06200 * ROP1_Mass)
    ROP2_Bi_Rad     = float_ur_boat(0.06200 * ROP2_Mass)
    
    Ves_Bi_Rad      = float_ur_boat(0.46000 * Ves_Mass)
    
    Det_Bi_Rad      = float_ur_boat(90.0000 * Det_Mass)
    
    FC_HDPE_Co_Rad  = 0.07000 * FC_HDPE_Mass
    FC_Tef_Co_Rad   = 0.50000 * FC_Tef_Mass
    FC_Cu_Co_Rad    = 0.00000 * FC_Cu_Mass
#     FC_Co_Rad       = float_ur_boat(FC_HDPE_Co_Rad + FC_Tef_Co_Rad + FC_Cu_Co_Rad )
    FC_Co_Rad       = float_ur_boat(FC_HDPE_Co_Rad)
    
    ICS_Co_Rad      = float_ur_boat(0.00000 * ICS_Mass)
    
    Out1_Co_Rad     = float_ur_boat(0.00000 * Out1_Mass)
    Out2_Co_Rad     = float_ur_boat(0.00000 * Out2_Mass)
    
    ROP1_Co_Rad     = float_ur_boat(0.07000 * ROP1_Mass)
    ROP2_Co_Rad     = float_ur_boat(0.07000 * ROP2_Mass)
    
    Ves_Co_Rad      = float_ur_boat(2.75000 * Ves_Mass)
    Det_Co_Rad      = float_ur_boat(40.0000 * Det_Mass)
    
    FC_HDPE_K_Rad   = 4.16000 * FC_HDPE_Mass
    FC_Tef_K_Rad    = 6.80000 * FC_Tef_Mass
    FC_Cu_K_Rad     = 0.03100 * FC_Cu_Mass
#     FC_K_Rad        = float_ur_boat(FC_HDPE_K_Rad + FC_Tef_K_Rad + FC_Cu_K_Rad )
    FC_K_Rad        = float_ur_boat(FC_HDPE_K_Rad)
    
    ICS_K_Rad       = float_ur_boat(0.03100 * ICS_Mass)
    
    Out1_K_Rad      = float_ur_boat(0.03100 * Out1_Mass)
    Out2_K_Rad      = float_ur_boat(0.03100 * Out2_Mass)
    
    ROP1_K_Rad      = float_ur_boat(4.16000 * ROP1_Mass)
    ROP2_K_Rad      = float_ur_boat(4.16000 * ROP2_Mass)
    
    Ves_K_Rad       = float_ur_boat(0.09000 * Ves_Mass)
    Det_K_Rad       = float_ur_boat(700.000 * Det_Mass)
    
    FC_HDPE_Tl_Rad  = 0.00755 * FC_HDPE_Mass
    FC_Tef_Tl_Rad   = 0.57500 * FC_Tef_Mass
    FC_Cu_Tl_Rad    = 0.00431 * FC_Cu_Mass
#     FC_Tl_Rad       = float_ur_boat(FC_HDPE_Tl_Rad + FC_Tef_Tl_Rad + FC_Cu_Tl_Rad )
    FC_Tl_Rad       = float_ur_boat(FC_HDPE_Tl_Rad)
    
    ICS_Tl_Rad      = float_ur_boat(0.00431 * ICS_Mass)
    
    Out1_Tl_Rad     = float_ur_boat(0.00431 * Out1_Mass)
    Out2_Tl_Rad     = float_ur_boat(0.00431 * Out2_Mass)
    
    ROP1_Tl_Rad     = float_ur_boat(0.00755 * ROP1_Mass)
    ROP2_Tl_Rad     = float_ur_boat(0.00755 * ROP2_Mass)

    Ves_Tl_Rad      = float_ur_boat(0.04310 * Ves_Mass)
    Det_Tl_Rad      = float_ur_boat(80.0000 * Det_Mass)
    
    return (FC_Bi_Rad, FC_Co_Rad, FC_K_Rad, FC_Tl_Rad, 
            ICS_Bi_Rad, ICS_Co_Rad, ICS_K_Rad, ICS_Tl_Rad, 
            Out1_Bi_Rad, Out1_Co_Rad, Out1_K_Rad, Out1_Tl_Rad, 
            Out2_Bi_Rad, Out2_Co_Rad, Out2_K_Rad, Out2_Tl_Rad,
            ROP1_Bi_Rad, ROP1_Co_Rad, ROP1_K_Rad, ROP1_Tl_Rad, 
            ROP2_Bi_Rad, ROP2_Co_Rad, ROP2_K_Rad, ROP2_Tl_Rad, 
            Ves_Bi_Rad, Ves_Co_Rad, Ves_K_Rad, Ves_Tl_Rad, 
            Det_Bi_Rad, Det_Co_Rad, Det_K_Rad, Det_Tl_Rad)
```

```python
def array_maker_vol(csv_name):
    with open(csv_name, 'r') as csvfile:

        csvreader = csv.reader(csvfile)

        # This skips the first row of the CSV file.
        # csvreader.next() also works in Python 2.
        next(csvreader)

        for row in csvreader:            
            try: 
                vol_array = np.append(vol_array, [row], axis = 0)
            except: 
                vol_array = np.array([row])
                
    return vol_array
```

```python
def array_maker_rad(csv_name):
    with open(csv_name, 'r') as csvfile:

        csvreader = csv.reader(csvfile)

        # This skips the first row of the CSV file.
        # csvreader.next() also works in Python 2.
        next(csvreader)
        
        for row in csvreader:            
            try: 
                rad_array = np.append(rad_array, [row], axis = 0)
            except: 
                rad_array = np.array([row])
                
    return rad_array
```

```python
def b_finder(FC_RejR, ICS_RejR, Out_RejR, ROP_RejR, Ves_RejR, 
             FC_Rad , ICS_Rad , Out1_Rad, Out2_Rad, 
             ROP1_Rad, ROP2_rad, Ves_Rad, Det_Rad,
             KeV, Kg):
    
    
    FC_Fin  = (FC_Rad * FC_RejR*.001)
    ICS_Fin = (ICS_Rad * ICS_RejR*.001)
    Out_Fin = ((Out1_Rad + Out2_Rad) * Out_RejR*.001)
    ROP_Fin = (Det_Rad * ROP_RejR*2 *.001) #Temporary, will change with inclusion of PMT's
    Ves_Fin = (Ves_Rad * Ves_RejR*.001)
    
    b_Net   = (3.15*10**7)*(FC_Fin + ICS_Fin + Out_Fin + Ves_Fin + ROP_Fin)
#     b_Net   = (3.15*10**7)*(FC_Fin + ICS_Fin + Out_Fin + Ves_Fin)
    
    
    b       = (b_Net) / (Kg * KeV)
    
    return b
```
